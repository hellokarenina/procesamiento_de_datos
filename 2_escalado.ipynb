# Escalado
Conjunto de datos: iris (incluido en sklearn)
Este conjunto de datos es compacto y perfecto para demostrar técnicas de escalado.

```python
# Importar librerías necesarias
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar el conjunto de datos
iris = load_iris(as_frame=True)
data = iris.data
```

```python
# Visualizar las estadísticas iniciales
print("Estadísticas iniciales:")
data.describe()
```

```python
# Aplicar MinMaxScaler
minmax_scaler = MinMaxScaler()
data_minmax = pd.DataFrame(minmax_scaler.fit_transform(data), columns=data.columns)
```

```python
# Aplicar StandardScaler
standard_scaler = StandardScaler()
data_standard = pd.DataFrame(standard_scaler.fit_transform(data), columns=data.columns)
```

### 1. ¿Qué diferencias encuentras entre las estadísticas originales y las del escalado MinMax?


```python
print("Estadísticas originales:")
print(data.describe())

print("\nEstadísticas después de MinMaxScaler:")
print(data_minmax.describe())
```

En los datos originales, las características tienen diferentes rangos, valores mínimos, máximos, medias y desviaciones estándar. Después de aplicar MinMaxScaler, todas las características quedan escaladas entre 0 y 1, conservando la distribución relativa de los datos. El valor mínimo de cada columna pasa a ser 0 y el máximo 1, mientras que la media y la desviación estándar también cambian de escala.

### 2. Crea una gráfica de la distribución de la variable 'sepal length (cm)' antes y después del escalado.

```python
plt.figure(figsize=(10, 6))

plt.hist(
    data['sepal length (cm)'],
    bins=20,
    alpha=0.6,
    label='Datos originales'
)

plt.hist(
    data_minmax['sepal length (cm)'],
    bins=20,
    alpha=0.6,
    label='Después de MinMaxScaler'
)

plt.title('Distribución de sepal length antes y después del escalado')
plt.xlabel('Sepal length')
plt.ylabel('Frecuencia')
plt.legend()

plt.show()
```

### 3. Crea una gráfica de dispersión entre 'sepal length (cm)' y 'sepal width (cm)' con los datos escalados por StandardScaler.

```python
plt.figure(figsize=(8, 6))

plt.scatter(
    data_standard['sepal length (cm)'],
    data_standard['sepal width (cm)'],
    alpha=0.7
)

plt.title('Sepal length vs Sepal width con StandardScaler')
plt.xlabel('Sepal length (estandarizado)')
plt.ylabel('Sepal width (estandarizado)')

plt.show()
```

StandardScaler puede ser más adecuado para el conjunto de datos Iris porque estandariza las características para que tengan una escala comparable, con media cercana a 0 y desviación estándar cercana a 1. Esto evita que algunas variables tengan mayor influencia por su escala y puede mejorar el desempeño de algoritmos sensibles a las distancias.
